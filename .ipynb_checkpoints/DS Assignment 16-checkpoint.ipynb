```python
import pandas as pd
import seaborn as sns
```

```python
data = pd.read_csv('CarBuyers.csv')
```

```python
data.head()
```

```python
data=data.loc[(data['Fuel']=='petrol')| (data['Fuel']=='diesel')]
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
from sklearn.model_selection import train_test_split
```

```python
data_train, data_test = train_test_split(data, test_size=0.3, stratify=data['Fuel'], random_state=42)
print(data_train.shape, data_test.shape)
```

```python
features= ['Power']
dt_classification = DecisionTreeClassifier(max_depth = 3) # Increase max_depth to see effect in the plot
dt_classification.fit(data_train[features], data_train['Fuel'])
```

```python
def calculate_accuracy(predictions, actuals):
    if(len(predictions) != len(actuals)):
        raise Exception("The amount of predictions did not equal the amount of actuals")
    
    return (predictions == actuals).sum() / len(actuals)
```

```python
predictionsOnTrainset = dt_classification.predict(data_train[features])
predictionsOnTestset = dt_classification.predict(data_test[features])

accuracyTrain = calculate_accuracy(predictionsOnTrainset, data_train.Fuel)
accuracyTest = calculate_accuracy(predictionsOnTestset, data_test.Fuel)

print("Accuracy on training set " + str(accuracyTrain))
print("Accuracy on test set " + str(accuracyTest))
```

```python
from sklearn import tree
import graphviz

def plot_tree_classification(model, features, class_names):
    # Generate plot data
    dot_data = tree.export_graphviz(model, out_file=None, 
                          feature_names=features,  
                          class_names=class_names,  
                          filled=True, rounded=True,  
                          special_characters=True)  

    # Turn into graph using graphviz
    graph = graphviz.Source(dot_data)  

    # Write out a pdf
    graph.render("decision_tree")

    # Display in the notebook
    return graph 
```

```python
plot_tree_classification(dt_classification, features, data.Fuel.unique())
```
