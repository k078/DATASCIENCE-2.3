```python
import pandas as pd
import seaborn as sns
from sklearn.cluster import KMeans
```

```python
ps = sns.load_dataset('penguins')
```

```python
ps.head()
```

```python
sns.pairplot(ps)
```

Ik denk bij 5 verschillende grafieken clusters te zien. Het zijn gescheiden groepen die beide dicht bij elkaar liggen.

```python
ps = ps.dropna()
```

```python
features = ['bill_length_mm','bill_depth_mm','flipper_length_mm', 'body_mass_g']
km = KMeans(n_clusters=3, random_state=42).fit(ps[features])
```

```python
ps['cluster'] = km.predict(ps[features])
```

```python
ps.head()
```

```python
ps.cluster.value_counts()
```

```python
from sklearn import metrics
from sklearn.metrics import pairwise_distances
```

```python
metrics.silhouette_score(ps[features], km.labels_, metric='euclidean')
```

```python
sns.pairplot(ps, hue="cluster")
```

```python
contingency_table = ps.groupby(['species','cluster']).size().unstack('cluster', fill_value=0)
contingency_table
```

Wat we uit deze tabel kunnen concluderen is dat de penguins ongeveer allemaal dezelfde eigenschappen hebben en niet erg uiteenlopen. De enige uitloop is dat de Gentoo's in cluster 2 de enige soort zijn en in dat cluster dus aanzienlijk verschillen ten opzichte van de andere soorten. Daardoor zijn de Gentoo's in cluster '1' ondervertegenwoordigd.
